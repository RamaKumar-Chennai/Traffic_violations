```python
# Creating a new column Violation_Category from the  original column "Description"

def categorize(desc: str) -> str:
    desc = str(desc).lower()
    if "speed" in desc:
        return "Speeding"
    elif "registration" in desc:
        return "Registration"
    elif "license" in desc:
        return "License"
    elif "insurance" in desc:
        return "Insurance"
    elif "seatbelt" in desc:
        return "Seatbelt"
    elif "alcohol" in desc:
        return "Alcohol"
    elif "unregistered" in desc:
        return "Unregistered"
    else:
        return "Other"

# ✅ Add new column to your existing df
df["Violation_Category"] = df["Description"].apply(categorize)

# Optional: make it efficient
df["Violation_Category"] = df["Violation_Category"].astype("category")

print(df.head())
print(df.dtypes)
```
